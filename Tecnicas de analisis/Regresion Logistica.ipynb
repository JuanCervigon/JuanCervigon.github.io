# Librerías


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Simular datos

```python
# Simular dos grupos en dos dimensiones
sig=1.5
x11=np.random.normal(7,sig,250)
x21=np.random.normal(2,sig,250)
y1=np.zeros(250)
x12=np.random.normal(2,sig,250)
x22=np.random.normal(6,sig,250)
y2=np.ones(250)

# Unir los puntos en un array X de dimensión 500 x 2 y dejar el vector y con 0 y 1
x1=np.concatenate((x11,x12),axis=0)
x2=np.concatenate((x21,x22),axis=0)
y=np.concatenate((y1,y2),axis=0).T # Vector (500x1) con 250 ceros y 250 unos
X=np.transpose([x1,x2]) # Matriz con 500 puntos (500x2) para clasificar



```

```python
# Visualizar los datos X,y

fig, ax=plt.subplots()
ax.plot(x11,x21,'o')
ax.plot(x12,x22,'o')
ax.set(xlabel='Eje x1', ylabel='Eje x2');
```

# Regresion Logistica

```python
# Estimar el modelo

from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression(random_state = 0) 
lr.fit(X,y)

y_pred = lr.predict(X)

```

```python
from sklearn.metrics import confusion_matrix 
mc = confusion_matrix(y, y_pred) 
  
print ("Matriz de confusion: \n", mc)
```

```python
# calculo de las betas
b0,b1,b2=lr.intercept_[0],lr.coef_[0,0],lr.coef_[0,1]

# calculo del hiperplano de separacion
yh=(-b0-b1*x1)/b2
```

```python
# Grafico

# Visualizar los datos X,y con la recta de separacion x1,yh

fig, ax=plt.subplots()
ax.plot(x11,x21,'o')
ax.plot(x12,x22,'o')
ax.plot(x1,yh)
ax.set(xlabel='Eje x1', ylabel='Eje x2');
```

# Otros

```python
# Grafico 1

from matplotlib.colors import ListedColormap 

X1, X2 = np.meshgrid(np.arange(start = X[:, 0].min() - 1,  
                               stop = X[:, 0].max() + 1, step = 0.01), 
                     np.arange(start = X[:, 1].min() - 1,  
                               stop = X[:, 1].max() + 1, step = 0.01)) 
  
plt.contourf(X1, X2, lr.predict( 
             np.array([X1.ravel(), X2.ravel()]).T).reshape( 
             X1.shape), alpha = 0.75, cmap = ListedColormap(('red', 'green'))) 
  
plt.xlim(X1.min(), X1.max()) 
plt.ylim(X2.min(), X2.max()) 
  
for i, j in enumerate(np.unique(y)): 
    plt.scatter(X[y == j, 0], X[y == j, 1], 
                c = ListedColormap(('red', 'green'))(i), label = j) 
      
plt.title('Regresión logística') 
plt.xlabel('x1') 
plt.ylabel('x2') 
plt.legend() 
plt.show()
```
