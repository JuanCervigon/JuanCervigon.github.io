# Librerías


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Datos

```python
# Simular cuatro clusters en dos dimensiones
sig=1.5
x1=np.random.normal(2,sig,250)
y1=np.random.normal(2,sig,250)
x2=np.random.normal(2,sig,250)
y2=np.random.normal(6,sig,250)
x3=np.random.normal(6,sig,250)
y3=np.random.normal(2,sig,250)
x4=np.random.normal(6,sig,250)
y4=np.random.normal(6,sig,250)

# Unir los puntos en un array X de dimensión 1000 x 2
x=np.concatenate((x1,x2,x3,x4),axis=0)
y=np.concatenate((y1,y2,y3,y4),axis=0)
X=np.transpose([x,y])


```

```python
# Visualizar los datos X

fig, ax=plt.subplots()
ax.plot(x1,y1,'o')
ax.plot(x2,y2,'o')
ax.plot(x3,y3,'o')
ax.plot(x4,y4,'o')
ax.set(xlabel='Eje x', ylabel='Eje y')
plt.show()
```

# K-Means

```python
# Crear los clusters

from sklearn.cluster import KMeans
modelo = KMeans(n_clusters=4)
modelo.fit(X)

grupos = modelo.predict(X) # Vector que contiene el cluster al que pertenece cada observación
```

```python
df=pd.DataFrame(X,columns=['x','y'])
df['cluster']=grupos
```
